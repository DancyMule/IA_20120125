```python
import twint
import pandas as pd 
import matplotlib.pyplot as plt
import nest_asyncio


nest_asyncio.apply()
```

```python
c = twint.Config()
#c.Username ="JuanJoseRosales"
c.Search = "JuanJoseRosales"
c.Limit = 100
c.Pandas = True

twint.run.Search(c)
```

```python
df = twint.storage.panda.Tweets_df
df
```

```python
tweet = df['tweet']
tweet
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
%config IPCompleter.greedy=True
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
count_vectorizer = CountVectorizer(stop_words = 'spanish')
```

```python
count_vectorizer = CountVectorizer()
```

```python
matrix = count_vectorizer.fit_transform(tweet)
```

```python
data_matrix = matrix.todense()
```

```python
frecuencias = pd.DataFrame(data_matrix, columns = count_vectorizer.get_feature_names())
```

```python
frecuencias
```

```python
salida = cosine_similarity (frecuencias, frecuencias)
```

```python
salida
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,10))
plt.imshow(salida)
plt.colorbar()
```
